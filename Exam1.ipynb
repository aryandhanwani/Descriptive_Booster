```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
df=pd.read_csv("household_dataset.csv")
```

```python
df.head()
```

```python
#identifying which columns are categoricl and numerical 
df.info()
```

```python
#Central Tendancy

income=df["Household_Income"]
age=df["Age_of_Household_Head"]

print(f"Mean of Income : {income.mean()}")
print(f"Median of Income : {income.median()}")
print(f"Mode of Income : {income.mode()[0]}")
print()
print(f"Mean of Age : {age.mean()}")
print(f"Median of Age : {age.median()}")
print(f"Mode of Age : {age.mode()[0]}")
```

```python
#Measure of Dispersion
df["Household_Income"].describe()
```

```python
#IQR
IQR = df["Household_Income"].quantile(0.75) - df["Household_Income"].quantile(0.25)
print("IQR:", IQR)
```

```python
#Distribution
sns.histplot(df["Household_Income"],bins=30,kde=True)
plt.title("Household Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()


print(f"Skewness : {skew(income)}")
print(f"kurtosis : {kurtosis(income)}")
```

```python
#Comaparing income by Education Level 
sns.boxplot(x=df["Household_Income"],y=df["Education_Level"])
plt.title("Income by Education Level")
plt.show()
```

```python
#Comaparing income by urban and rural 
sns.boxplot(x=df["Household_Income"], y=df["Urban_Rural"])
plt.title("Income: Urban vs Rural")
plt.show()
```
